```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, classification_report, ConfusionMatrixDisplay, confusion_matrix, roc_auc_score, RocCurveDisplay
```

EDA process of DATA

```python
df=pd.read_csv("/content/customer_churn_dataset-training-master.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Churn'].value_counts()
```

```python
# Converting gender object type to numeric
df['Gender'] = df['Gender'].map({'Female': 0, 'Male': 1})
```

```python
df['Gender'].value_counts()
```

```python
df['Subscription Type'].value_counts()
```

```python
# Converting subscription object type category to numeric value
df = pd.get_dummies(df, columns=['Subscription Type'])
print(df)
```

```python
df.head()
```

```python
df['Contract Length'].value_counts()
```

```python
# Converting Contract Length object type category to numeric value
df = pd.get_dummies(df, columns=['Contract Length'])
print(df)
```

```python
df.head()
```

```python
df = df.astype(int)
```

```python
df.head()
```

```python
df=df.drop(['CustomerID'],axis=1)
```

```python
df.head()
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.head()
```

```python
x = df.drop(['Churn'], axis=1)
y = df['Churn']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
# for classification
model = RandomForestClassifier(
    n_estimators=100,
    max_depth=5,
    class_weight="balanced",
    random_state=42)
model.fit(X_train_scaled, y_train)
```

```python

# for classification
model_XGB = XGBClassifier(objective="binary:logistic")
model_XGB.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
y_pred
```

```python
y_pred_XGB = model_XGB.predict(X_test_scaled)
y_pred_XGB
```

```python
# for classification
print("Accuracy RandomForest :", accuracy_score(y_test, y_pred))
print("Precision Random Forest:", precision_score(y_test, y_pred))
print("Recall Random Forest  :", recall_score(y_test, y_pred))
print("F1 Score Random Forest:", f1_score(y_test, y_pred))

print("\nClassification Report Random Forest:\n")
print(classification_report(y_test, y_pred))
```

```python

# for classification
print("Accuracy XGBoost :", accuracy_score(y_test, y_pred_XGB))
print("Precision XGBoost:", precision_score(y_test, y_pred_XGB))
print("Recall XGBoost  :", recall_score(y_test, y_pred_XGB))
print("F1 Score Random Forest:", f1_score(y_test, y_pred_XGB))

print("\nClassification Report XGBoost:\n")
print(classification_report(y_test, y_pred_XGB))
```

```python
def predict_churn(customer_data):
  customer_data_scaled = scaler.transform(pd.DataFrame([customer_data]))
  prediction = model.predict(customer_data_scaled)
  print("Prediction:", prediction[0])
  return True if prediction[0] == 1 else False
```
